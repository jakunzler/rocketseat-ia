# Resolução do Desafio de Estatística Descritiva

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dicionário de faturamento
dict_sales = {
    'date_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'value': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

## Exercício do módulo

Com base nas aulas do módulo de Estatística, utilize os métodos da biblioteca Pandas para tratar o dicionário abaixo e:

- Trazer a média das vendas
- Criar um gráfico de barras vertical mostrando o mês de referência e o valor
- Criar um gráfico de linhas mostrando o mês de referência e o valor.

```python
df_sales = pd.DataFrame(dict_sales)
df_sales
```

```python
# Convertendo a coluna date_ref para o tipo datetime
df_sales['date_ref'] = pd.to_datetime(df_sales['date_ref'])
df_sales
```

```python
# Calculando as estatísticas das vendas

# Calculando a média das vendas
mean_sales = df_sales['value'].mean()
print(f'Média de vendas: {mean_sales}')

# Calculando a mediana das vendas
median_sales = df_sales['value'].median()
print(f'Mediana de vendas: {median_sales}')

# Calculando a moda das vendas
mode_sales = df_sales['value'].mode()
print(f'Moda de vendas: {mode_sales}')

# Calculando o desvio padrão das vendas
std_sales = df_sales['value'].std()
print(f'Desvio padrão de vendas: {std_sales}')

# Calculando a variância das vendas
var_sales = df_sales['value'].var()
print(f'Variância de vendas: {var_sales}')

# Calculando o coeficiente de variação das vendas
cv_sales = std_sales / mean_sales
print(f'Coeficiente de variação de vendas: {cv_sales}')
```

```python
# Gráfico de barras das vendas
df_sales.plot(kind='bar', x='date_ref', y='value')
plt.xlabel('Data de Referência')
plt.ylabel('Valor de Vendas')
plt.title('Faturamento por Data de Referência')
plt.xticks(rotation=75)
plt.show()

# Gráfico de linha das vendas
plt.plot(df_sales['date_ref'], df_sales['value'])
plt.xlabel('Data de Referência')
plt.ylabel('Valor de Vendas')
plt.title('Faturamento por Data de Referência')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gráfico de barras e linha das vendas
fig, ax = plt.subplots()
ax.bar(df_sales['date_ref'], df_sales['value'])
ax.plot(df_sales['date_ref'], df_sales['value'], color='r', marker='o')
plt.xlabel('Data de Referência')
plt.ylabel('Valor de Vendas')
plt.title('Faturamento por Data de Referência em 2023')
plt.xticks(rotation=45)
plt.show()
```
