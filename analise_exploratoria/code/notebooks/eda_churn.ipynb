## 📘 **Resumo de Comandos Principais do `pandas`**

```python
import pandas as pd
```

---

### 📂 **Leitura de Dados**

```python
df = pd.read_csv('arquivo.csv')          # CSV
df = pd.read_excel('arquivo.xlsx')       # Excel
df = pd.read_json('arquivo.json')        # JSON
df = pd.read_sql(query, conexao)         # SQL
df = pd.read_html('url ou arquivo.html') # Tabelas HTML
```

---

### 🧾 **Criação de DataFrame**

```python
df = pd.DataFrame({
    'nome': ['Ana', 'Bruno'],
    'idade': [23, 35]
})
```

---

### 🔍 **Inspeção de Dados**

```python
df.head()             # Primeiras 5 linhas
df.tail()             # Últimas 5 linhas
df.info()             # Resumo do DataFrame
df.describe()         # Estatísticas descritivas
df.shape              # (linhas, colunas)
df.columns            # Nomes das colunas
df.index              # Índices do DataFrame
```

---

### 🎯 **Seleção e Filtros**

```python
df['coluna']                      # Seleção de uma coluna
df[['col1', 'col2']]              # Seleção de múltiplas colunas
df.loc[linha, 'coluna']           # Seleção por rótulo
df.iloc[0, 1]                     # Seleção por índice

df[df['coluna'] > 10]            # Filtro condicional
df.query('coluna == "valor"')    # Filtro com query
```

---

### ✏️ **Manipulação de Dados**

```python
df['nova'] = df['col1'] + df['col2']     # Nova coluna
df.drop('coluna', axis=1)                # Remover coluna
df.drop(0, axis=0)                       # Remover linha
df.rename(columns={'velho': 'novo'})     # Renomear coluna
df.sort_values('coluna', ascending=False) # Ordenar

df.fillna(0)             # Preencher valores nulos
df.dropna()              # Remover valores nulos
df.duplicated()          # Identificar duplicatas
df.drop_duplicates()     # Remover duplicatas
```

---

### 📊 **Estatísticas**

```python
df['coluna'].mean()      # Média
df['coluna'].sum()       # Soma
df['coluna'].min()       # Mínimo
df['coluna'].max()       # Máximo
df['coluna'].value_counts()  # Contagem de valores únicos
```

---

### 🔁 **Agrupamento e Agregação**

```python
df.groupby('coluna').mean()         # Média por grupo
df.groupby(['col1', 'col2']).sum()  # Agregação multi-index
df.pivot_table(index='A', columns='B', values='C', aggfunc='sum')  # Tabela dinâmica
```

---

### 🔗 **Mesclagem e Junções**

```python
pd.concat([df1, df2])                            # Concatenar DataFrames
pd.merge(df1, df2, on='id', how='inner')         # Merge por coluna
```

---

### 🕒 **Datas e Horários**

```python
df['data'] = pd.to_datetime(df['data'])          # Converter para datetime
df['ano'] = df['data'].dt.year                   # Extrair ano
df['mes'] = df['data'].dt.month                  # Extrair mês
```
