### üîç **Inspe√ß√£o de Dados**


---

### üéØ **Sele√ß√£o e Filtros**

```python
df['coluna']                      # Sele√ß√£o de uma coluna
df[['col1', 'col2']]              # Sele√ß√£o de m√∫ltiplas colunas
df.loc[linha, 'coluna']           # Sele√ß√£o por r√≥tulo

```

---

### ‚úèÔ∏è **Manipula√ß√£o de Dados**

```python
df['nova'] = df['col1'] + df['col2']     # Nova coluna
df.drop('coluna', axis=1)                # Remover coluna
df.drop(0, axis=0)                       # Remover linha
df.rename(columns={'velho': 'novo'})     # Renomear coluna
df.sort_values('coluna', ascending=False) # Ordenar

df.fillna(0)             # Preencher valores nulos
df.dropna()              # Remover valores nulos
df.duplicated()          # Identificar duplicatas
df.drop_duplicates()     # Remover duplicatas
```

---

### üìä **Estat√≠sticas**

```python
df['coluna'].mean()      # M√©dia
df['coluna'].sum()       # Soma
df['coluna'].min()       # M√≠nimo
df['coluna'].max()       # M√°ximo

```

---

### üîÅ **Agrupamento e Agrega√ß√£o**

```python

df.pivot_table(index='A', columns='B', values='C', aggfunc='sum')  # Tabela din√¢mica
```
